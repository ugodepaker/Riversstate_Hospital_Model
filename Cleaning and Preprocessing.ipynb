```python
####Read Raw File



import pandas as pd
import numpy as np
pd.set_option("display.max.columns", None)

df = pd.read_excel("C:/Users/USER/Downloads/hospital_data.xlsx")
```

```python
#convert columns to lower case
df.columns= df.columns.str.lower()
```

```python
###Drop columns
data = df.drop(['lgas','year','east','north', 'long', 'lat', 'northings', 'eastings'], axis=1)
```

```python
data.head()
```

```python
data.shape
```

```python
####Find out variables that have missing and null columns
```

```python
na = data.isnull().sum()
na_df = na.reset_index()
na_df.columns = ["Variables", "Mising_values_counts"]
```

```python
na_df
```

```python
df_na = na_df['Mising_values_counts'] > 0
vals = na_df[df_na]
vals
```

```python
missing_val = np.array(vals['Variables'])
missing_df = df[missing_val]
```

```python
missing_df
```

```python
####Replace missing columns with mean values for each columns
```

```python
a = [i for i in missing_df.columns]
mean_val = []
for i in a:
    l = round((missing_df[i].sum())/missing_df.shape[0])
    mean_val.append(l)
# meanx = mean_val[0]
# meanx
np.array(mean_val)
```

```python
dict_missing = dict(zip(a,mean_val))
```

```python
dict_missing
```

```python
dta = pd.DataFrame.from_dict(dict_missing, orient="index")
dta.columns = ["Mean_values"]
```

```python
dta["Mean_values"]
```

```python
for l, k in dict_missing.items():
    missing_df[l] = missing_df[l].fillna(k)
```

```python
missing_df.head(10)
```

```python
### merge columns togther
```

```python
l = df[['lgas','year']]
```

```python
pz = pd.concat([l, missing_df], axis=1)
```

```python
####check if there are missing values after cleaning
```

```python
# calculate outcome/class
# - This is done by getting the percentage of individual variables for each columns
# - Get the percentage mean accross each column
# - Define RANGE of the outcome by substracting maximum value from the minimum value..here the class can be decided
```

```python
ldata = []
for i in missing_df.columns:
    ldata.append(i)
```

```python
ldata
```

```python
sum_var = []
for i in ldata:
    b = missing_df[i].sum()
    sum_var.append(b)
```

```python
sum_var
```

```python
data_dict = missing_df.to_dict("list")
```

```python
data_dict
```

```python
arr_dict = []
for p, l in data_dict.items():
    arr_dict.append(l)
print(arr_dict)
```

```python
b_val =[]
for a in arr_dict:
    h_val =[]
    for b in a:
        p = round((b/sum(a))*100,2)
        h_val.append(p)
    b_val.append(h_val)
```

```python
b_val
```

```python
df_vals_p = pd.DataFrame(b_val)
```

```python
df_vals_p.head()
```

```python
df_vals_p.shape
```

```python
zp = []
lp = []
dict_p = df_vals_p.to_dict("list")
for i, n in dict_p.items():
    zp.append(i)
    lp.append(n)
```

```python
len(lp)
```

```python
n = []
for i in lp:
    j = round(sum(i)/124,2)
    n.append(j)
weight = pd.DataFrame(n, columns=["value_Weight"])
```

```python
#d_loc = df[['east','north', 'long', 'lat', 'northings', 'eastings']]
researched_score = 0.9
```
